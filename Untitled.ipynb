```python
import numpy as np

import keras

import matplotlib.pyplot as plt
```

```python
Nparam = 100

C = np.random.rand(Nparam)
D = np.random.rand(Nparam)

Nx = 50
x = np.arange(0,Nx)

y = np.zeros( [Nparam,Nx] )
for i in range(Nparam):
    y[i,:] = np.sin(x*C[i]+D[i])
    
training_data_input = np.array([C,D]).T #np.expand_dims( np.array([C,D]).T, -1 )
training_data_output = np.copy(y)
```

```python
from keras import backend as K
K.set_image_data_format('channels_last')

from keras.layers import Dense
from keras.models import Sequential

model = Sequential()

input_shape = (1,1)

model.add( Dense( 2, input_shape=(2,), activation='relu', name='Dense-1' ) )
model.add( Dense( 10, activation='relu', name='Dense-2' ) )
model.add( Dense( 25, activation='relu', name='Dense-3' ) )
model.add( Dense( 50, activation='linear', name='Out' ) )

model.summary(line_length=120) 

### model compilations
model.compile( loss='mean_squared_error',
               optimizer='RMSprop' )

history = model.fit( training_data_input, training_data_output,
                     epochs=10000,
                     batch_size=20,
                     verbose=True )
```

```python
Ntest = 100

C_test = np.random.rand(Ntest)
D_test = np.random.rand(Ntest)

y_test = np.zeros( [Nparam,Nx] )
for i in range(Nparam):
    y_test[i,:] = np.sin(x*C_test[i]+D_test[i])
    
test_data_input = np.array([C_test,D_test]).T #np.expand_dims( np.array([C,D]).T, -1 )
test_data_output = np.copy(y_test)
```

```python
y_emul = model.predict( test_data_input )
```
